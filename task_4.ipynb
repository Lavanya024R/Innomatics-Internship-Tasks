Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected!")
        return False
    else:
        print("Recharge Successful!")
        return True


while True:
    amt = int(input("Enter recharge amount: "))
    if validate_recharge(amt):
        break
    print("Please try again.\n")
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert!")
        else:
            print(f"{product}: Stock OK")


products = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Oil": 12
}

check_inventory(products)
```

Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)
    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare


while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(km, peak_hour)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_eligibility(attendance)
print("Status:", status)
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit & special character)"


pwd = input("Enter password: ")
print(check_password(pwd))
```
